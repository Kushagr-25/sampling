```python
import pandas as pd
data = pd.read_csv('creditcard.csv')
```

```python
data['Class'].value_counts()
```

```python
class_0 = data[data['Class'] == 0]
class_1 = data[data['Class'] == 1]
```

```python
import imblearn
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

features = data.columns

x_class_0 = class_0[features]
y_class_0 = class_0['Class']

x_class_1 = class_1[features]
y_class_1 = class_1['Class']

smote = SMOTE()

x_smote, y_smote = smote.fit_resample(pd.concat([x_class_0, x_class_1]), pd.concat([y_class_0, y_class_1]))

print('Original dataset shape:', Counter(pd.concat([y_class_0, y_class_1])))
print('Resampled dataset shape:', Counter(y_smote))
```

```python
x_smote
```

```python
import numpy as np

np.random.seed(0)
sample_size = 400

random_sample = x_smote.sample(n=sample_size, random_state=0)
random_sample
```

```python
import math
n = len(x_smote)
k = int(math.sqrt(n))

systematic_sample = x_smote.iloc[::k]
systematic_sample
```

```python
stratified_sample = x_smote.groupby('Class').apply(
    lambda x: x.sample(frac=0.001)
)
stratified_sample = stratified_sample.droplevel(0)
stratified_sample
```

```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import random

num_clusters = 5000
kmeans = KMeans(n_clusters = num_clusters, n_init = 'auto', random_state =42)

clusters = kmeans.fit_predict(x_smote)
clusters = pd.Series(clusters)

selected_clusters = random.sample(range(num_clusters), 5)
cluster_sample = x_smote.loc[clusters.isin(selected_clusters)]
cluster_sample
```

```python
import numpy as np

np.random.seed(0)
sample_size = 400

bootstrap_sample = x_smote.sample(n=sample_size, random_state=0, replace = True)
bootstrap_sample
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X = random_sample.drop(columns=['Class'])
y = random_sample['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

logreg_model = LogisticRegression(max_iter=1200)

logreg_model.fit(X_train, y_train)

y_pred = logreg_model.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```
